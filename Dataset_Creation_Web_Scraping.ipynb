```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
r = requests.get('https://en.wikipedia.org/wiki/Toy_Story_3')

#convert to beautiful soup object
soup = bs(r.content)

content = soup.prettify()
print(content)
```

## 1. Scrape Infobox

```python
## Get infobox

info_box = soup.find(class_ = "infobox vevent")

print(info_box.prettify())
```

```python
info_rows = info_box.find_all("tr")
for row in info_rows:
  print(row.prettify())
```

```python
# we see that table head <th> has key and table data <td> has value for python dict
movie_info = {}

for i, row in enumerate(info_rows):
  if i == 0:
    movie_info["Title"] = row.find("th").get_text()
  elif i!=1:
    key = row.find("th").get_text()
    val = row.find("td").get_text()
    movie_info[key] = val
```

```python
for row in info_rows:
  if row.find("li"):
    print(row.find_all("li"))
    # print(row.find("li").get_text())
    # print([i.get_text() for i in row.find_all("li")])
```

```python
#handle multiple vals for key in list format
def get_content_value(row):
  if row.find("li"):
    return [li.get_text().replace("\xa0", " ") for li in row.find_all("li")]
  else:
    return row.find("td").get_text()

movie_info2 = {}

for i, row in enumerate(info_rows):
  if i == 0:
    movie_info2["Title"] = row.find("th").get_text(" ", strip = True)
  elif i!=1:
    key = row.find("th").get_text(" ", strip = True)
    val = get_content_value(row)
    if type(val) != list:
      movie_info2[key] = val.replace("\xa0", " ")
    else:
      movie_info2[key] = val
```

## 2. Scrape infobox for all movies in Disney Films

```python
r = requests.get("https://en.wikipedia.org/wiki/List_of_Walt_Disney_Pictures_films")

soup = bs(r.content)

contents = soup.prettify()

print(contents)
```

```python
table = soup.select(".wikitable.sortable i")
```

```python
table[-10:]
```

```python
for row in table:
  if row.a:
    print(row.a["href"], ",", row.get_text())
```

```python
def get_content_value(row):
  if row.find("li"):
    return [li.get_text().replace("\xa0", " ") for li in row.find_all("li")]
  else:
    return row.find("td").get_text()

def get_info_box(url):
  r = requests.get(url)
  soup = bs(r.content)

  infobox = soup.find(class_ = "infobox vevent")
  movie = {}

  if infobox:
    info_rows = infobox.find_all("tr")

    for i, row in enumerate(info_rows):
      if row and row.find("th") and row.find("td"):
        if i == 0:
          movie["Title"] = row.find("th").get_text(" ", strip = True)
        elif i!=1:
          key = row.find("th").get_text(" ", strip = True)
          val = get_content_value(row)
          if type(val) != list:
            movie[key] = val.replace("\xa0", " ")
          else:
            movie[key] = val

  return movie
```

```python
r = requests.get("https://en.wikipedia.org/wiki/List_of_Walt_Disney_Pictures_films")
soup = bs(r.content)

#pick all tables
contents = soup.select(".wikitable.sortable i")

movie_list = []
for row in contents:
  if row.a:
    url = row.a["href"]
    temp = get_info_box("https://en.wikipedia.org/" + url)
    if len(temp) != 0:
      movie_list.append(temp)
```

## 3. Saving into JSON File

```python
import json

def save_data(title, data):
  with open(title, 'w', encoding = 'utf-8') as f:
    json.dump(data, f, ensure_ascii = False, indent = 2)
```

```python
save_data("disney_data.json", movie_list)
```
